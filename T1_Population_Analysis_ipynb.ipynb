```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['figure.figsize'] = (11,6)
plt.rcParams['figure.constrained_layout.use'] = True
main_csv = '/content/API_SP.POP.TOTL_DS2_en_csv_v2_38144 .csv'
meta_country_csv = '/content/Metadata_Country_API_SP.POP.TOTL_DS2_en_csv_v2_38144.csv'
meta_indicator_csv = '/content/Metadata_Indicator_API_SP.POP.TOTL_DS2_en_csv_v2_38144.csv'
df = pd.read_csv(main_csv, skiprows=4)
meta_country = pd.read_csv(meta_country_csv)
meta_indicator = pd.read_csv(meta_indicator_csv)
print("Main shape:", df.shape)
df.head()
```

```python
year_cols = [c for c in df.columns if c.strip().isdigit()]
year_cols = sorted(year_cols, key=lambda x: int(x))
print("Year columns found:", len(year_cols), "first..last ->", year_cols[0], "...", year_cols[-1])
latest_year = year_cols[-1]
latest_year
```

```python
print("Total countries (rows):", df['Country Name'].nunique())
missing_by_col = df.isnull().sum().sort_values(ascending=False).head(10)
print("Top missing counts:\n", missing_by_col)
```

```python
top5 = df[['Country Name', latest_year]].dropna().sort_values(by=latest_year, ascending=False).head(5)
bottom5 = df[['Country Name', latest_year]].dropna().sort_values(by=latest_year, ascending=True).head(5)
print("Top 5:\n", top5.to_string(index=False))
print("\nBottom 5:\n", bottom5.to_string(index=False))
```

```python
top20 = df[['Country Name', latest_year]].dropna().sort_values(by=latest_year, ascending=False).head(20)
plt.figure(figsize=(10,8))
plt.barh(top20['Country Name'][::-1], top20[latest_year][::-1])
plt.xlabel("Population")
plt.title(f"Top 20 Countries by Population ({latest_year})")
plt.show()
plt.savefig('/content/top20_population.png', bbox_inches='tight')
```

```python
vals = pd.to_numeric(df[latest_year], errors='coerce').dropna()
plt.figure(figsize=(10,6))
plt.hist(vals, bins=30)
plt.title(f"Distribution of Country Populations ({latest_year})")
plt.xlabel("Population")
plt.ylabel("Number of Countries")
plt.show()
plt.savefig('/content/population_histogram.png', bbox_inches='tight')
```

```python
if '1960' in year_cols:
    df['pop_1960'] = pd.to_numeric(df['1960'], errors='coerce')
    df['pop_latest'] = pd.to_numeric(df[latest_year], errors='coerce')
    df['Growth_%_1960_to_latest'] = ((df['pop_latest'] - df['pop_1960']) / df['pop_1960']) * 100
    growth_sample = df[['Country Name','1960', latest_year, 'Growth_%_1960_to_latest']].dropna().sort_values('Growth_%_1960_to_latest', ascending=False).head(10)
    growth_sample
else:
    print("1960 not present â€” skipping growth calc.")
```

```python
if 'Growth_%_1960_to_latest' in df.columns:
    fastest = df[['Country Name', 'Growth_%_1960_to_latest']].dropna().sort_values('Growth_%_1960_to_latest', ascending=False).head(10)
    fastest
else:
    print("Growth percentage not computed (missing 1960).")
```

```python
if len(year_cols) >= 2:
    prev_year = year_cols[-2]
    df[f'YoY_change_{prev_year}_to_{latest_year}'] = pd.to_numeric(df[latest_year], errors='coerce') - pd.to_numeric(df[prev_year], errors='coerce')
    df[['Country Name', prev_year, latest_year, f'YoY_change_{prev_year}_to_{latest_year}']].dropna().sort_values(f'YoY_change_{prev_year}_to_{latest_year}', ascending=False).head(10)
else:
    print("Not enough yearly columns for YoY.")
```

```python
countries = ['India', 'China', 'United States', 'Indonesia', 'Brazil']
years = year_cols

plt.figure(figsize=(12,6))
for country in countries:
    r = df[df['Country Name'] == country]
    if r.empty:
        print("Missing country:", country)
        continue
    vals = r[years].T.squeeze().astype(float)
    plt.plot(years, vals, marker='o', label=country)

plt.xticks(rotation=45)
plt.xlabel("Year")
plt.ylabel("Population")
plt.title("Population Over Time (selected countries)")
plt.legend()
plt.show()

plt.savefig('/content/selected_countries_time_series.png', bbox_inches='tight')
```

```python
try:
    import seaborn as sns
    yr_df = df[years].apply(pd.to_numeric, errors='coerce').dropna(how='all')
    corr = yr_df.corr()
    plt.figure(figsize=(12,8))
    sns.heatmap(corr, cmap='coolwarm', linewidths=0.3)
    plt.title("Correlation between years (population)")
    plt.show()
    plt.savefig('/content/years_correlation_heatmap.png', bbox_inches='tight')
except Exception as e:
    print("Seaborn heatmap skipped or failed:", e)
```

```python
region_col_candidates = [c for c in meta_country.columns if 'region' in c.lower() or 'Region' in c]
print("Possible region columns:", region_col_candidates)
for candidate in ['Region', 'region', 'Region Name', 'RegionName']:
    if candidate in meta_country.columns:
        region_col = candidate
        break
else:
    region_col = None

if region_col:
    merged = df.merge(meta_country[[ 'Country Code', region_col]], on='Country Code', how='left')
    region_sum = merged.groupby(region_col)[latest_year].sum().dropna().sort_values(ascending=False)
    plt.figure(figsize=(12,6))
    region_sum.plot(kind='bar')
    plt.title(f"Population by {region_col} ({latest_year})")
    plt.ylabel("Population")
    plt.xticks(rotation=45)
    plt.show()
    plt.savefig('/content/region_population.png', bbox_inches='tight')
else:
    print("No region-like column found in meta_country; skip region plot.")

```
